```python
import pandas as pd
import plotly_express as px
```

```python
outages = pd.read_csv("../DataAcquisition/outages.csv")
```

```python
fig = px.density_mapbox(outages, lat='latitude', lon='longitude', radius=20, center=dict(lat=outages.latitude.mean(), lon = outages.longitude.mean()), zoom=10, mapbox_style='open-street-map')
fig.show()
```
